```python
# Assume a 30 x 30 grid. Plot these gridlines in matplotlib.
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-15, 15)
y=np.arange(-15, 15)
for i in range(-15,15):
    plt.plot(np.full((30, ), i), y, color="yellow")
    plt.plot(x, np.full((30, ), i), color="yellow")
plt.grid()    
```

```python
#Shear Transformation using loops
import matplotlib.pyplot as plt
import numpy as np 
shear = np.array([[1, 1],[0,1]])
for i in range(-15,15):
    for j in range(-15,15):
        sm=shear.dot(np.array([i,j]))
#       print(sm)
        plt.scatter(sm[0], sm[1])
plt.grid()        
```

```python
#Parallelized approach without using loops

gmatrix = np.zeros((2,900)) # Creating Grid for 900 points
n = 0 

for i in range(-15, 15): 
    for j in range(-15, 15): 
        # Appending values
        gmatrix[0][n] = i
        gmatrix[1][n] = j
        n += 1

print(gmatrix)

gmatrix = shear.dot(gmatrix)
for i in range(900):
    plt.plot(gmatrix[0][i], gmatrix[1][i], ".g")
```
